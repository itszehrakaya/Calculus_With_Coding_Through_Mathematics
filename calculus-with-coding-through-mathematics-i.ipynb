# FUNCTIONS I

- A function $f$ is a rule that assigns to each element x in a set **A** exactly one element, called $f(x)$, in a set **B**.

- The set **A** is called the _domain of the function_. The range of $f$ is the set of all possible values of $f(x)$ as x varies throughout the domain.

- The domain of a function denotes the set of all possible inputs that the function takes in, while the range specifies the set of all possible outputs.

- The symbol that represents an arbitrary number in the domain of a function $f$ is called an _independent variable_. The symbol that represents a number in the range of $f$ is called a _dependent variable_:

So if we write $y = f(x)$, then **x** is the _independent variable_ and **y** is the _dependent variable_.

- If **x** is in the domain of the function $f$, then when **x** enters the machine, it is accepted as an input and the machine produces an output $f(x)$ according to the rule of the function.

# Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from warnings import filterwarnings
```

```python
filterwarnings("ignore",category=UserWarning)
```

```python
plt.style.use("dark_background")
plt.rcParams["figure.autolayout"] = True
```

```python
class RESULTS:
    pass
```

```python
def PLOT_FUNCTION(x,y,lineon:bool=False)->RESULTS:
    fig = plt.figure(figsize=(15,8))
    ax = fig.add_subplot(1,1,1)
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    #ax.spines['right'].set_position('zero')
    #ax.spines['top'].set_position('zero')
    ax.spines['right'].set_color('white')
    ax.spines['top'].set_color('white')
    plt.title("Plot of Function")
    if lineon:
        plt.axvline(x=0.3,
               color="r",
               label='axvline - full height')
        plt.plot(x,y)
        plt.tight_layout()
        plt.show()
    else:
        plt.plot(x,y)
        plt.tight_layout()
        plt.show()
```

```python
def PLOT_CIRCLE(x_point:int or float,
                y_point:int or float,
                r:int or float,
                lineon:bool=False)->RESULTS:
    fig = plt.figure(figsize=(15,8))
    ax = fig.add_subplot(1,1,1)
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    #ax.spines['right'].set_position('zero')
    #ax.spines['top'].set_position('zero')
    ax.spines['right'].set_color('white')
    ax.spines['top'].set_color('white')
    c_one = plt.Circle((x_point,y_point),
                           r,
                           color="green",
                           fill=False,
                           alpha=0.9,
                           lw=3)
    ax = plt.gca()
    plt.title("UNIT CIRCLE")
    ax.add_patch(c_one)
    ax.set_aspect('equal',
                      adjustable='datalim')
    if lineon:
        plt.axvline(x=5.3,
                   color="r",
                   label='axvline - full height')
        ax.plot()
        plt.tight_layout()
        plt.show()
    else:
        ax.plot()
        plt.tight_layout()
        plt.show()
```

# Example of Function and Output

* $f(x) = x^2 + 4$

```python
def FUNCTION_ONE(x)->RESULTS:
    y = (x**2)+4
    return y
```

```python
y:int = FUNCTION_ONE(3)
print(f"Result of function: {y}")
```

# Vertical Line Test

- Not every curve you draw is the graph of a function. A function $ƒ$ can have only one value $ƒ(x)$ for each **x** in its domain, so _no vertical line_ can intersect the graph of a function more than once. For instance, a circle cannot be the graph of a function since some vertical lines inter-sect the circle twice.

```python
def FUNCTION_TWO(x)->RESULTS:
    y = (1-(x)**2)**1/2
    return y
```

- $ƒ(x) = \sqrt{1-x^2}$ 

```python
x = np.linspace(-1,1,200)
y = []
for i_ in x:
    y.append(FUNCTION_TWO(i_))
y = np.array(y)
```

```python
PLOT_FUNCTION(x,y,lineon=True)
```

```python
PLOT_CIRCLE(5,5,1,lineon=True)
```

# Piecewise-Defined Functions

- Sometimes a function is described by using different formulas on different parts of its domain. One example is the **absolute value function**.

```python
def FUNCTION_PIECEWISE(x)->RESULTS:
    if x >= 0:
        return x
    elif x < 0:
        return -x
```

$$
f(x)=
\begin{cases}
x & \quad \text{$x >= 0$}\\ 
-x & \quad \text{$x < 0$}
\end{cases}
$$

```python
y = FUNCTION_PIECEWISE(1)
print(f"RESULT OF FUNCTION: {y}")
```

```python
y = FUNCTION_PIECEWISE(-2)
print(f"RESULT OF FUNCTION: {y}")
```

```python
x = np.linspace(-1,2,200)
y = []
for i_ in x:
    y.append(FUNCTION_PIECEWISE(i_))
```

```python
PLOT_FUNCTION(x,y)
```
